```python
import numpy as np

# function to clean data
def clean(data):
    data = data.copy()

    for item in features:
        data = data.replace(np.inf, np.nan)
        
    return data

df = clean(df)
```

```python
from sklearn.model_selection import train_test_split

# train-val-test split
my_train, my_test = train_test_split(df, test_size=0.2, random_state=42) 
my_train, my_val = train_test_split(my_train, test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# removing the target from the data, setting features
y_train = my_train[target]
X_train = my_train[features]
y_val = my_val[target]
X_val = my_val[features]
y_test = my_test[target]
X_test = my_test[features]
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

# Feature importance & permutation importance

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
imputer = SimpleImputer()

X_train_imputed = imputer.fit_transform(X_train_scaled)
X_val_imputed = imputer.transform(X_val_scaled)
```

```python
from xgboost import XGBRegressor

xgb_eli5_model = XGBRegressor(
    n_estimators=183
)

xgb_eli5_model.fit(X_train_imputed, y_train)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

permuter = PermutationImportance(
    xgb_eli5_model,
    scoring='neg_mean_absolute_error',
    n_iter=5,
    random_state=42,
)

permuter.fit(X_val_imputed, y_val)
```

```python
permuter.feature_importances_
feature_names = X_val.columns.tolist()
pd.set_option('max_rows', None)
pd.Series(permuter.feature_importances_, feature_names).sort_values(ascending=False)
```

```python
eli5.show_weights(
    permuter,
    top=None,
    feature_names=feature_names
)
```

## Changing the features to remove useless/harmful features

```python
minimum_importance = 0
mask = permuter.feature_importances_ > minimum_importance
features = X_train.columns[mask]
X_train = X_train[features]
X_val = X_val[features]
X_test = X_test[features]
```

```python
X_train.columns
```

## final model

```python
xgb_pipeline = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler(),
    XGBRegressor(learning_rate=0.12103,
                max_depth=5,
                n_estimators=358,
                reg_alpha=0.73132)
)

xgb_pipeline.fit(X_train, y_train)
```

```python
xgb_pred_train = xgb_pipeline.predict(X_train)
xgb_pred_val = xgb_pipeline.predict(X_val)
xgb_pred_test = xgb_pipeline.predict(X_test)

print('Train r2 score: ', r2_score(y_train, xgb_pred_train))
print('Train MAE: ', mean_absolute_error(y_train, xgb_pred_train), '\n')
print('Validation r2 score: ', r2_score(y_val, xgb_pred_val))
print('Validation MAE: ', mean_absolute_error(y_val, xgb_pred_val), '\n')
print('Test r2 score: ', r2_score(y_test, xgb_pred_test))
print('Test MAE: ', mean_absolute_error(y_test, xgb_pred_test))
```

```python
import shap

row = X_val.loc[[250]]

explainer = shap.TreeExplainer(xgb_shap_model)
row_processed = impute_scale_pipeline.transform(row)

shap_values = explainer.shap_values(row_processed)

shap.initjs()
shap.force_plot(
    base_value=explainer.expected_value,
    shap_values=shap_values,
    features=row
)
```
